```python
def print_array_info(arra):
    color = 'green'
    print('Array:', colored(arra, color))
    print('Size:', colored(arra.size, color), 'equals length', colored(len(arra), color))
    print('Shape:', colored(arra.shape, color))
    print('Data type:', colored(arra.dtype, color))
    print('Dimention count:', colored(arra.ndim, color))
```

### Some math functions

```python
def parabola2(x):
    return x ** 2

def parabola3(x):
    return x ** 3

def parabola05(x):
    return x ** 0.5

def tan(x):
    return np.tan(x)

def tanh(x):
    return np.tanh(x)

def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)

def linear(x):
    return -x

def linear2(x):
    return x

def linear0(x):
    return 0

def linear1(x):
    return 1

```

```python
def func(start, count, step, fn):
    x = []
    y = []
    stop = count * step + start + 1

    for i in np.arange(start, stop, step):
        x.append(i)
        y.append(fn(i))
    
    return np.array(x), np.array(y)
```

### Generate metrics

#### Mean Absolute Error (MAE)

```python
def mae(actual, prediction):
    return np.absolute((actual - prediction))
```

#### Mean Squared Error (MSE)

```python
def mse(actual, prediction):
    return np.square(actual - prediction)
```

#### Y-hat

```python
def yhat(x, y):
    k, d = np.polyfit(x, y, 1)
    y_pred = k*x + d
    return y_pred
```

#### Apply Y-hat, MAE and MSE

```python
sin_yhat = yhat(y1, y0)
cos_yhat = yhat(y2, y0)
tanh_yhat = yhat(y3, y0)
sqrt_yhat = yhat(y4, y0)


# print('sin(x)', sin_yhat)
print('mean sin(x):', sin_yhat.mean())

# print('cos(x)', cos_yhat)
print('mean cos(x):', cos_yhat.mean())

# print('tanh(x)', tanh_yhat)
print('mean tanh(x):', tanh_yhat.mean())

# print('sqrt(x)', sqrt_yhat)
print('mean sqrt(x):', sqrt_yhat.mean())


fig, ax = plt.subplots()

ax.plot(x1, sin_yhat)
ax.plot(x2, cos_yhat)
ax.plot(x3, tanh_yhat)
ax.plot(x4, sqrt_yhat)
```

```python
# print(x1)
# print(x2)
# print(x3)
# print(x4)

sin_mae = mae(y1, y0)
cos_mae = mae(y2, y0)
tanh_mae = mae(y3, y0)
sqrt_mae = mae(y4, y0)


# print('sin(x)', sin_mae)
print('mean sin(x):', sin_mae.mean())

# print('cos(x)', cos_mae)
print('mean cos(x):', cos_mae.mean())

# print('tanh(x)', tanh_mae)
print('mean tanh(x):', tanh_mae.mean())

# print('sqrt(x)', sqrt_mae)
print('mean sqrt(x):', sqrt_mae.mean())


fig, ax = plt.subplots()

ax.plot(x1, sin_mae)
ax.plot(x2, cos_mae)
ax.plot(x3, tanh_mae)
ax.plot(x4, sqrt_mae)
```

```python
sin_mse = mse(y1, y0)
cos_mse = mse(y2, y0)
tanh_mse = mse(y3, y0)
sqrt_mse = mse(y4, y0)


# print('sin(x)', sin_mse)
print('mean sin(x):', sin_mse.mean())

# print('cos(x)', cos_mse)
print('mean cos(x):', cos_mse.mean())

# print('tanh(x)', tanh_mse)
print('mean tanh(x):', tanh_mse.mean())

# print('sqrt(x)', sqrt_mse)
print('mean sqrt(x):', sqrt_mse.mean())


fig, ax = plt.subplots()

ax.plot(x1, sin_mse)
ax.plot(x2, cos_mse)
ax.plot(x3, tanh_mse)
ax.plot(x4, sqrt_mse)
```
